```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
import os




```

**Άσκηση 1 — Φόρτωση Δεδομένων**

Φορτώστε το αρχείο Excel. Τα δεδομένα περιλαμβάνουν:

• Χαρακτηριστικά: Στήλες A έως K (Δείκτες απόδοσης & δραστηριότητας).

• Στόχος (Target): Στήλη Status (1=Healthy, 2=Bankrupt).

• Αγνοήστε τη στήλη "Year"

```python
df = pd.read_excel('Dataset2Use_Assignment.xlsx') 
df.head() #5 πρώτες εγραφές
```

```python
df.info() #10716 συνολικές εγγραφές
```

```python
df = df.drop(columns=['ΕΤΟΣ']) #Διώχνουμε το  Έτος
df.head()
```

```python
counts = df['ΕΝΔΕΙΞΗ ΑΣΥΝΕΠΕΙΑΣ (=2) (ν+1)'].value_counts()
print("Counts:\n", counts)
ratio = counts[1] / counts[2]
print(f"\nRatio (Healthy/Bankrupt): {ratio:.2f}")
```

### Άσκηση 2 — Προεπεξεργασία
1. Κανονικοποίηση: Χρησιμοποιήστε StandardScaler στα χαρακτηριστικά.

```python
scaler = StandardScaler()
X = df.drop(columns=['ΕΝΔΕΙΞΗ ΑΣΥΝΕΠΕΙΑΣ (=2) (ν+1)'])
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X.head()
```

2. Κωδικοποίηση Στόχου: Μετατρέψτε το Status σε δυαδική μορφή (0=Healthy,
1=Bankrupt).

```python
df = df.rename(columns={'ΕΝΔΕΙΞΗ ΑΣΥΝΕΠΕΙΑΣ (=2) (ν+1)': 'status'})
df['status'] = df['status'].map({1: 0, 2: 1})
```

3. Διαχωρισμός: Δημιουργήστε Train/Test sets (π.χ. test_size=0.25, stratified).

```python
y = df['status']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
```

**Άσκηση 3 — Έλεγχος Ισορροπίας (Class Balancing)**

Ελέγξτε την αναλογία Healthy:Bankrupt στο Training Set. Αν οι υγιείς επιχειρήσεις είναι
περισσότερες από 3 φορές τις χρεωκοπημένες (αναλογία > 3:1), κάντε downsampling (τυχαία
αφαίρεση) στην κλάση Healthy ώστε η αναλογία να γίνει ακριβώς 3:1.

```python

#Γνωρίζουμε η ότι η αναλογία Healthy:Bankrupt είναι περίπου 42.21:1, άρα πρέπει να κάνουμε downsampling στην κλάση Healthy.

X_healthy = X_train[y_train == 0]
y_healthy = y_train[y_train == 0]
X_bankrupt = X_train[y_train == 1]
y_bankrupt = y_train[y_train == 1]
    
X_healthy_downsampled, y_healthy_downsampled = resample(
        X_healthy, y_healthy,
        replace=False,    
        n_samples=3 *(y_train == 1).sum(), 
        random_state=42
    )
    
X_train = pd.concat([X_healthy_downsampled, X_bankrupt])
y_train = pd.concat([y_healthy_downsampled, y_bankrupt])
    

print(f"Νέα Αναλογία Healthy: {(y_train == 0).sum()}, Bankrupt: {(y_train == 1).sum()}")
print(f"Ratio μετά το downsampling: {(y_train == 0).sum() / (y_train == 1).sum():.2f}")
```

**Άσκηση 4 — Εκπαίδευση Μοντέλων**

Εκπαιδεύστε και αξιολογήστε τους παρακάτω ταξινομητές:
• Linear Discriminant Analysis (LDA)
• Logistic Regression (max_iter=1000)
• K-Nearest Neighbors (KNN) (δοκιμάστε k=5, 7, 9)
• Naïve Bayes (GaussianNB)
• Support Vector Machines (SVM) (RBF kernel)
• Neural Networks (MLP) (64, 32 nodes, max_iter=400)

```python
models = {
    "LDA": LinearDiscriminantAnalysis(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "KNN (k=5)": KNeighborsClassifier(n_neighbors=5),
    "KNN (k=7)": KNeighborsClassifier(n_neighbors=7),
    "KNN (k=9)": KNeighborsClassifier(n_neighbors=9),
    "Naïve Bayes": GaussianNB(),
    "SVM (RBF)": SVC(kernel='rbf'),
    "MLP (64, 32)": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=400, random_state=42)
}

# Train and evaluate
print(f"{'Model':<30} {'Accuracy':<10} {'Recall (Bankrupt)':<20} {'Specificity (Healthy)':<20} {'Constraints Met'}")
print("="*100)

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    report = classification_report(y_test, y_pred, output_dict=True)
    recall_bankrupt = report['1']['recall']      # Sensitivity
    specificity = report['0']['recall']          # Specificity (Recall of Healthy class)
    acc = accuracy_score(y_test, y_pred)
    
    constraints_met = "YES" if recall_bankrupt >= 0.62 and specificity >= 0.70 else "NO"
    
    print(f"{name:<30} {acc:.4f}     {recall_bankrupt:.4f}               {specificity:.4f}               {constraints_met}")
```

### Αξιολόγηση Αποτελεσμάτων

**Γενική Παρατήρηση:**
Παρατηρούμε ότι παρόλο που η **ακρίβεια (Accuracy)** είναι υψηλή σε όλα τα μοντέλα (85% - 91%), αυτό είναι παραπλανητικό λόγω της ανισορροπίας των κλάσεων στο Test Set (3:1). Τα μοντέλα τείνουν να προβλέπουν την πλειοψηφούσα κλάση (Healthy).

**Ανάλυση Περιορισμών:**
Κανένα από τα αρχικά μοντέλα δεν ικανοποίησε ταυτόχρονα τους περιορισμούς:
*   **Recall (Bankrupt) ≥ 62%**
*   **Specificity (Healthy) ≥ 70%**

Το καλύτερο Recall πέτυχε το **KNN (k=9)** με 54.84%, το οποίο όμως απέχει από το στόχο του 62%. Το Specificity είναι εξαιρετικά υψηλό σε όλα τα μοντέλα (>85%), που σημαίνει ότι τα μοντέλα είναι πολύ "συντηρητικά" στο να προβλέψουν χρεωκοπία.

**Επόμενο Βήμα:**
Έχουμε ήδη κάνει downsampling στο training set φέρνοντας την αναλογία στο **3:1** (Healthy:Bankrupt), θα δοκιμάσουμε να ενισχύσουμε περαιτέρω την ικανότητα εντοπισμού χρεωκοπιών χρησιμοποιώντας **βάρη**. Θα δώσουμε μεγαλύτερη "ποινή" στα λάθη που αφορούν τις χρεωκοπημένες επιχειρήσεις, ώστε να αναγκάσουμε τα μοντέλα να αυξήσουν το Recall τους, θυσιάζοντας πιθανώς λίγο από το πολύ υψηλό Specificity.

**Άσκηση 5 — Αξιολόγηση & Αρχείο Εξόδου**

Δημιουργήστε το αρχείο "classification_results.xlsx" στον φάκελο output. Για κάθε μοντέλο καταγράψτε τις εξής στήλες:
• Classifier Name
• Train/Test Set
• Samples Count
• TP (True Positives)
• TN
• FP
• FN
• Precision
• Recall
• F1 Score
• Accuracy

```python
# Define models with class weights
models_weighted = {
    "Logistic Regression (Balanced)": LogisticRegression(max_iter=1000, class_weight='balanced'),
    "SVM (Balanced)": SVC(kernel='rbf', class_weight='balanced'),
    "Logistic Regression (Custom 1:3)": LogisticRegression(max_iter=1000, class_weight={0: 1, 1: 3}),
    "SVM (Custom 1:3)": SVC(kernel='rbf', class_weight={0: 1, 1: 3}),
}

print("Training with Class Weights to meet constraints (Recall >= 62%, Specificity >= 70%)...\n")
print(f"{'Model':<35} {'Accuracy':<10} {'Recall (Bankrupt)':<20} {'Specificity (Healthy)':<20} {'Constraints Met'}")
print("="*105)

results = []

for name, model in models_weighted.items():
    model.fit(X_train, y_train)
    
    # Evaluate on both Train and Test sets
    for set_name, X_set, y_set in [("Train", X_train, y_train), ("Test", X_test, y_test)]:
        y_pred = model.predict(X_set)
        
        # Calculate metrics
        tn, fp, fn, tp = confusion_matrix(y_set, y_pred).ravel()
        precision = precision_score(y_set, y_pred)
        recall = recall_score(y_set, y_pred)
        f1 = f1_score(y_set, y_pred)
        acc = accuracy_score(y_set, y_pred)
        
        results.append({
            "Classifier Name": name,
            "Train/Test Set": set_name,
            "Samples Count": len(y_set),
            "TP (True Positives)": tp,
            "TN": tn,
            "FP": fp,
            "FN": fn,
            "Precision": precision,
            "Recall": recall,
            "F1 Score": f1,
            "Accuracy": acc
        })
        
        # Print only Test results for the console output requested previously
        if set_name == "Test":
            report = classification_report(y_set, y_pred, output_dict=True)
            recall_bankrupt = report['1']['recall']     
            specificity = report['0']['recall']         
            constraints_met = "YES" if recall_bankrupt >= 0.62 and specificity >= 0.70 else "NO"
            
            print(f"{name:<35} {acc:.4f}     {recall_bankrupt:.4f}               {specificity:.4f}               {constraints_met}")

# Create DataFrame and save to Excel
import os
if not os.path.exists('output'):
    os.makedirs('output')

results_df = pd.DataFrame(results)
results_df.to_excel('output/classification_results.xlsx', index=False)
print("\nResults saved to output/classification_results.xlsx")
```
